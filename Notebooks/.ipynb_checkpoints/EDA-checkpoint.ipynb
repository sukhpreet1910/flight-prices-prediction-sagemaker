### 7.2 Date of Journey 

```python
eda_helper_functions.dt_univar_plots(train, var = 'date_of_journey', target = 'price')
```

```python
(
    train.groupby(pd.Grouper(key = 'date_of_journey', freq = 'M'))
    ['price'].mean()
    .to_frame()
    .set_axis(['March', 'April', 'May', 'June'], axis = 0)
    .rename_axis('months')
    .plot(
        kind = 'bar',
        color = '#B22222FF',
        figsize = (6, 4)
    )
)
plt.xticks(rotation = 35)
plt.show()
```

```python
(
    train.groupby(pd.Grouper(key = 'date_of_journey', freq = 'W'))
    ['price']
    .mean()
    .to_frame()
    .rename_axis('Weeks')
    .plot(
        kind = 'bar',
        color = '#B22222FF',
        figsize = (6, 4)
    )
)
plt.xticks(rotation = 35)
plt.show()
```

```python
(
    train
    .groupby([pd.Grouper(key = 'date_of_journey', freq = 'M'), 'source'])
    ['price']
    .mean()
    .unstack()
    .set_axis(['March', 'April', 'May', 'June'])
    .rename_axis('Month')
    .plot(
        kind = 'bar',
        figsize = (10, 4),
        edgecolor = 'Black',
        colormap = 'Set2'
    )
)




plt.xticks(rotation = 35)
plt.show()
```

### 7.3 Departure Time 

```python
train.dep_time
```

```python
(
    train
    .assign
    (dep_time = lambda df_ : 
        (
        np.select([df_['dep_time'].dt.hour.between(4, 12, inclusive='left'),
                   df_['dep_time'].dt.hour.between(12, 16, inclusive = 'left'),
                   df_['dep_time'].dt.hour.between(6, 20, inclusive = 'left')
                  ], 
                  ['Morning', 'AfterNoon', 'Evening'], 
                  default='Night')
        )
     )
    .pipe(eda_helper_functions.num_cat_bivar_plots, 'price', 'dep_time')
)
```

```python
train.head()
```

### 7. 4 Arrival Time

```python
(
    train
    .assign(arrival_time = lambda df_:
            (
                np.select([df_['arrival_time'].dt.hour.between(4, 12, inclusive = 'left'),
                           df_['arrival_time'].dt.hour.between(12, 16, inclusive = 'left'),
                           df_['arrival_time'].dt.hour.between(16, 20, inclusive = 'left')],
                         ['Morning', 'Afternoon', 'Evening'],
                         default = 'night')
            )
           )
    .pipe(eda_helper_functions.num_cat_bivar_plots, 'price', 'arrival_time')
)
```
